## Univariate Analysis

#### We will start off with importing our libraries

```python
import pandas as pandaforman                # this is for data manipulation
import numpy as numpyforops                 # this is for numerical operations
import matplotlib.pyplot as mplforplotting  # this is for plotting
import seaborn as seabornforvis             # this is for visualisations
```

#### Data loading and overview

```python
df = pandaforman.read_csv(r"D:\Downloads\Cleaned_Churn_modelling.csv")
df.head()      #allows us a preview into our dataset     
```

```python
df.describe()  #provides us with insight regarding basic statistical figures
```

```python
df.isnull().sum() #this is to check for the presence of missing values
                  #as this is a cleaned dataset, the results below prove the absence of missing values
```

### Numerical Variables
#### We will start our analysis by delving into our categorical variables

```python
numerical_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
for col in numerical_cols:
    mplforplotting.figure(figsize=(12, 5))
    mplforplotting.subplot(1, 2, 1)
    seabornforvis.histplot(df[col], kde=True)
    mplforplotting.title(f'{col} Distribution')

    mplforplotting.subplot(1, 2, 2)
    seabornforvis.boxplot(x=df[col])
    mplforplotting.title(f'{col} Boxplot')

    mplforplotting.tight_layout()
    mplforplotting.show()
```

## Categorical Variables
#### Now, we shift our focus to the categorical variables

```python
categorical_cols = ['Gender', 'HasCrCard', 'IsActiveMember']

for col in categorical_cols:
    mplforplotting.figure(figsize=(6, 4))
    seabornforvis.countplot(data=df, x=col)
    mplforplotting.title(f'{col} Countplot')
    mplforplotting.show()
    # for gender, 0 represents males while 1 represents females
    # for the other two visualisations, 0 means no while 1 means yes
```

## Geography
#### Lastly, we will analyse our three geographies

```python
geo_cols = ['Geography_France', 'Geography_Germany', 'Geography_Spain']
df[geo_cols].sum().plot(kind='bar')
mplforplotting.title('Geography Distribution')
mplforplotting.ylabel('Count')
mplforplotting.xticks(rotation=0)
mplforplotting.show()
```
